##Setting 

Logarithmic distribution is a popular distribution in insurance study to model insurance claim frequency.

The PMF of logarithmic distribution is as follows:

$$PMF(x)=\frac{-p^x}{xln(1-p)}$$, where $x=1, 2, 3, \dots, \infty$. 

The mean of the distribution is equal to $\frac{-1}{ln(1-p)}\frac{p}{1-p}$
The variance of the distribution is equal to $-\frac{p^2+p ln(1-p)}{(1-p)^2(ln(1-p))^2}$

##Q1 [6 Points]

* Fill in the function to use the direct method to compute the PMF for $x=1, \dots, k$. This computation needs to be based on array operations. No loops are allowed.

* Assume that p=0.8, run the function constructed to return the PMF for $x=1, \dots, 20$.

```python
def LogDist_direct (p,k):
  #Fill in the operations here
    x=np.arange(1,k+1)
    PMF=-p**x/x/np.log(1-p)
    return PMF
```

```python
##Call the function to return PMF corresponding to x=1, ..., 20
#print out the PMF array result
LogDist_direct (p=0.8,k=20)
```

##Q2 [8 Points]

* Fill in the function to use the iterative method to compute the PMF for $x=1, \dots k$

* Assume that p=0.8, run the function constructed to return the PMF for $x=1, \dots, 20$.


```python
def LogDist_iterative (p, k):
  #Fill in the operations here
    PMF_temp=-p/np.log(1-p)
    PMF=np.empty(k)
    PMF[0]=PMF_temp
    for i in range(2,k+1):
        PMF_temp=PMF_temp*p*(i-1)/i
        PMF[i-1]=PMF_temp
    return PMF
```

```python
##Call the function to return PMF corresponding to x=1, ..., 20 and p=0,8
#print out the PMF array result
LogDist_iterative (p=0.8, k=20)
```

## Q3 [6 Points]

* Construct a function to return the mean and variance of the distribution

* Print out the mean and variance of this distribution assuming $p=0.8$

```python
def LogDist_mv (p):
  #Fill in the operations here
  mean=-1/np.log(1-p)*p/(1-p)
  variance=-(p**2+p*np.log(1-p))/(1-p)**2/(np.log(1-p))**2
  return mean, variance
```

```python
##Call the function to return PMF corresponding to x=1, ..., 20 and p=0,8
#print out the PMF array result
LogDist_mv (0.8)
```

##Q4 [10 Points]

The expected value of a discrete distribution follows:

$E(x)=\sum_i x_i p_i$

For the distribution we work with, we need to sum up infinite terms. If we only sum up $x_i p_i$ from $i=1$ up to $i=k$, where $k$ is a large value, we can get a good approximation although not exact.

Construct an array that gives how much error we have for the expected value when we choose k=1, 2, ..., 30. 

```python
#construct the array here 
PMF=LogDist_direct (p=0.8,k=30)
x=np.arange(1,31)
Estimation=np.cumsum(PMF*x)
Error_array=Estimation-LogDist_mv(p=0.8)[0]
print(np.abs(Error_array))
```

##Setting 


A continuous distribution B has the following probability density function (PDF):

$f(x)=\frac{1}{150}t(t - 5) +\frac{2}{45}, \text{where }  0\leq x \leq 10$


##Q1 [6 Points]





Construct a function to do the following:
* Generate 1000 evenly spaced points between 0 and 20. 
* Compute the pdf corresponding to each point based on Boolean mask


```python
def continuous_pdf():
    #Fill in the operations here
    x=np.linspace(0,20,1000)
    pdf=1/150*x*(x-5)+2/45
    pdf[x>10]=0
    return pdf

#print the pdf array
continuous_pdf()
```

## Q2 [4 Points]

* Print out the average pdf for the points we sampled between 1.00001 and 1.02001 based on Boolean mask

```python
x=np.linspace(0,20,1000)
pdf_avg=continuous_pdf()[(x>=1.00001) * (x<=1.02001) ]
print(pdf_avg)
```
